```python
import matplotlib.pyplot as plt
import numpy as np
T = 32
a = []
b = []
p = 0
q =0
i= 0
e = 0
count =0
while (T != q):
  E = e
  e = 220 - T
  P = e - E 
  U = e*0.98+i*0.04664+P*0.9
  p = 0.2*U - 0.05*(T-32)
  q = T
  T = T + p
  count = count + 1
  i = i + e
  print(T)
  a.append(T)
  b.append(count)
y = np.array(a)
x = np.array(b)
plt.xlabel("Time")
plt.ylabel("Temperature")
print("total iterations to get steady state",count)
plt.plot(x,y)
plt.show()
```
